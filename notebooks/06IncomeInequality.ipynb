- S80/S20
- The Palma ratio
- poverty rate

- Definitions of Income Ineqaulity: OECD

> **S80/S20** is the ratio of the average income of the 20% richest to the 20% poorest; **P90/P10** is the ratio of the upper bound value of the ninth decile (i.e. the 10% of people with highest income) to that of the first decile; **P90/P50** of the upper bound value of the ninth decile to the median income; and **P50/P10** of median income to the upper bound value of the first decile. The **Palma ratio** is the share of all income received by the 10% people with highest disposable income divided by the share of all income received by the 40% people with the lowest disposable income.

```python
col_names = ['Country', 'Year', 'Decile 1 Income', 'Decile 2 Income',
       'Decile 3 Income', 'Decile 4 Income', 'Decile 5 Income',
       'Decile 6 Income', 'Decile 7 Income', 'Decile 8 Income',
       'Decile 9 Income', 'Decile 10 Income', 'Mean Income', 'Population']
decile_income = [name.split()[0]+name.split()[1] for name in col_names if len(name.split())==3]

col_names = col_names[:2] + decile_income + ['Mean_Income', 'Population']
col_names = [name.lower() for name in col_names]

data.columns = col_names
```

- 우크라이나 2004년 자료를 확인한다.

```python
data[data.country == "Ukraine"].loc[4602:4604]
```

- 2004년 Ukraine의 10분위 평균소득이 0으로 입력되어 있다.

- 해당 기록을 제거하는 방법이 있고 다른 imputed method로 채워 넣을 수도 있다. <br>
- 여기선 평균과 각 분위별 소득을 이용하여 10분위 소득을 정확히 계산할 수 있으므로 계산한 값을 사용하자.

$$y_{10} = 10\times \bar{y} - \sum^9_{i=1} y_i$$

```python
data.loc[4603, "decile10"] = 10*data.loc[4603, "mean_income"] - data.loc[4603, "decile1":"decile9"].sum()
data[4602:4605]
```

- 한국과 싱가포르의 10분위가 전체 표본에서 가장 높은 것은 확인히 필요하다. [상위 10%의 소득 비중](https://data.worldbank.org/indicator/SI.DST.10TH.10)

### 국가 이름 식별

- 대한민국 같은 경우 "Repulic of Korea", "Korea, Repubric", "South Korea" 등등 여러 가지 이름을 사용한다. 혼란을 피하기 위해 중요한 철자로 자료에서 사용한 국가명을 식별할 수 있도록 하자.


- 'Republic of the Niger'와 'Nigeria'를 구분하기 위해서는 전체 국가명을 정확히 입력하는 수 밖에 없다.

```python
country_list = data.country.unique()

def name(subs):
    country_name = list(filter(lambda x: subs.lower() in x.lower(), country_list))
    if subs in country_name: return subs
    else: return ", ".join(country_name)
    
print(name("dem"))
print(name("nig"))
```

```python
def plot_class_incomes(country, ax=None):
    
    incomes = data[data.country == country]
    incomes = incomes.set_index(incomes.year)
    
    if ax == None:
        incomes.iloc[:,2:-2].plot(legend=False, title=country)
    else:
        incomes.iloc[:,2:-2].plot(legend=False, ax=ax, title=country)

countries = [name("Korea"), "United States", 
             "Japan", "Taiwan", 
             "Malaysia", "Thailand",
             "Sweden", "China"]

fig, ax = plt.subplots(4, 2, figsize=(10,10))
ax = ax.flatten()

for i, country in enumerate(countries):
    plot_class_incomes(country, ax=ax[i])
    
plt.tight_layout()
```

## Lorenz 곡선


- Lorenz 곡선은 소득기준 하위 $x$ 비율의 사람들의 소득이 사회 총소득에서 차지하는 비율을 표시
- 개별소득 자료가 아니므로 각 분위에 포함된 사람들의 소득이 동일하다고 가정하고 Lorenz곡선을 그린다.

### 누적 소득 비중

- Lorenz 곡선은 소득비중의 단순한 누적합이므로 누적합으로 구성된 DataFrame을 따로 만든다.
- 꼭 필요하진 않지만 곡선의 원점을 0으로 하려면 `decile 0`을 추가해 준다.

```python
cum = panel.loc[:,"decile1":"decile10"].cumsum(axis=1)
lorenz = cum.div(cum.iloc[:,-1], axis=0)
lorenz.insert(0,"decile0", 0)
```

## Gini 계수

- Discrete한 10분위 자료이므로 각 소득분위의 사람들의 소득이 동일하다고 가정하므로 지니계수의 정의를 이용하여 계산한다.

$$\text{Gini coefficient} = \frac{1}{2} 
\frac{\sum_i \sum_{j\ne i} \left| y_i - y_j \right|}{n \left( n-1 \right) } 
\frac{1}{\bar{y}}=
\frac{\sum_i \sum_{j > i} \left| y_i - y_j \right|}{n \left( n-1 \right) } 
\frac{1}{\bar{y}}
$$


- 연속성을 가정하면 다음과 같이 계산할 수도 있다.
- 보통 교과서에서 많이 볼 수 있는 공식이다.

$$ \text{Gini coefficient} = \frac{A}{A+B} = 2A = 1 - 2B = 1 - 2 \int^1_0 L(x) dx $$

- Gini coefficient의 도형으로서의 성질 중 재미있는 사실은 기울기가 체증한다는 점이다.
- 따라서 1분위와 10분위의 소득비중에 크게 의존하게 된다.

- Internet 상에서 돌아다니는 방법들로 계산하면 다른 결과를 얻기도 한다.

```python
dist = lorenz.loc[(name('kor'), 1999)].to_numpy()
perfect = np.arange(0, 1.01,.1)

plt.bar(perfect, dist, width=.1, 
        facecolor='gainsboro', edgecolor='gray')
plt.bar(perfect, perfect, width=.1, alpha=.5,
        facecolor='lightgray', edgecolor='gray' )
plt.plot([0,1], [0,1], color='dimgray')
plt.vlines(x=1, ymin=0, ymax=1, color='dimgray')
plt.xlim([0,1.05]);
```

```python
income_dist = panel.loc[:,"decile1":"decile10"].copy()

def gini(country, year, income_dist=income_dist):
    
    dist = income_dist.loc[(country, year)].to_numpy().reshape(-1,1)
    n = dist.shape[0]
    m = np.mean(dist)
    sum_ = np.abs(np.triu(dist-dist.T)).sum()
    
    gini_coeff = sum_/(n*(n-1)*m)
    
    return gini_coeff
```

### Gini 계수의 의미

- 전체 구성원에서 무작위로 두 명을 뽑아 둘의 소득을 비교해 보면 상대적으로 소득이 많은 사람과 작은 사람의 소득 비율은 다음과 같이 계산할 수 있다.

$$\frac{\text{소득이 많은 사람의 소득}}{\text{소득이 낮은 사람의 소득}} = \frac{1+\text{Gini}}{1-\text{Gini}}$$

- Gini 계수가 1/3 이라면 위의 비율이 2이다.


- 2018년 대한민국의 Gini 계수는 0.34 였다.<br>
따라서 무작위로 선택한 두 명의 소득 비율은 평균 1.34/0.66 = 2.03배이다.
- Iceland의 2019년 Gini 계수가 0.25 이므로 그 평균 비율은 1.25/0.75 = 1.67배이다.


- 주의<br>
$E\left( \max\left( \frac{m_i}{m_j} , \frac{m_j}{m_i}\right) \vert m_i \right)$ 가 아니라 $E \left( \max\left( \frac{m_i}{m_j} , \frac{m_j}{m_i}\right) \right)$를 의미한다.




```python
x = np.arange(0.2, 0.82, 0.02)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, (1+x)/(1-x))
ax.plot([0.2, 0.33, 0.5, 0.6, 0.8], [1.5, 2, 3, 4, 9], "bo")
ax.set_xlabel('Gini Coefficient', fontsize=14)
ax.set_ylabel('Expected Ratio', fontsize=14)

ax.set_yticks(range(1,10))
ax.set_yticklabels(range(1,10))

ax.set_xticks([.2, .3, .33, .4, .5, .6, .7, .8])
ax.set_xticklabels([.2, .3, .33, .4, .5, .6, .7, .8])
ax.grid()

fig.suptitle("Statistical Property of Gini Coefficient", fontsize=16)
fig.tight_layout()
```
