### 시장균형

- 완전경쟁시장에선 수요와 공급이 균형가격과 균형량을 결정
  - 수요는 가격에 대한 조건부 구매 **계획**
  - 공급은 가격에 대한 조건부 생산 **계획**
  - 시장에서 관찰하는 가격과 거래량은 시장**균형**
<br>
$$\begin{array}{rcc}
\text{수요 :} & q_D & = & D(p) \\
\text{공급 :} & q_S & = & S(p) \\
\text{균형조건 :} & q_D & = & q_S
\end{array}$$
<br>
관찰할 수 있는 값은 '균형량' $q^*$ 와 '균형가격' $p^*$ 뿐이다.
<br><br>
- **수요계획**과 **공급계획** 추정에서 가장 큰 문제는 <br>
시장균형의 **변화 원인**이 아주 다양하고<br>
균형량과 균형가격은 시장에서 결정되는 **내생변수**라는 것이다.

### 추정에 필요한 시장이론


1. **완전경쟁시장**
  - 개별 소비자나 개별 공급자의 의사결정과정에서 가격은 주어진 것으로 생각한다.
  - 시장균형은 수요와 공급에 의해 결정된다.
<br><br>  
2. 공급과 수요의 식별
  - 시장에 충격이 없다면 공급과 수요는 식별이 불가능
  - 공급곡선을 식별하기 위해서는 수요곡선의 이동이 필수적이며
  - 수요곡선은 공급곡선의 변화를 통해 식별해야 한다.
  - 공급(수요)곡선을 추정하기 위해서는 해당 기간에 수요(공급)곡선이 충분히 변해야 한다.
<br><br>


3. Ceteris paribus, 가격 이외 변수들은 **고정**
  - 공급곡선을 추정하기 위해서는 수요곡선의 이동만을 고려해야 하므로<br>
  공급곡선을 이동시키는 요인들을 통제해야 한다.
  - 마찬가지로 수요곡선을 추정하기 위해서는 수요곡선에 영향을 미치는 요일들을 통제한다.
<br><br>

### 비율과 변환

- 사회과학분야를 비롯한 여러 분야에서 (특히 시간과 관련된) 변수의 절대적인 크기는 별 의미가 없다.
- 대부분 변화에 초점을 맞추어 모형을 만들고 분석한다.
- 대부분의 경우 절대적인 변화보다는 상대적인 변화를 주로 사용


- 상대적인 크기를 측정하는 가장 보편적인 방법은 비율이다.


$$ \frac{p_{t+1}}{p_t}$$

- 간단한 표기를 위해 로그변환을 취하기도 한다.

$$\log \frac{p_{t+1} }{p_t} = \log p_{t+1} - \log p_t$$

따라서 자연로그의 경우 로그값이 0.69면 수준에 관계없이 2배, 1.1이면 3배 증가한 값이다.

### 로그 변환의 장점

추정방정식에서 변수의 로그변환을 사용하는 경우가 많다. <br>
이분산 해결을 위해서도 사용하고, 특히 변수 변화량 사이 관계를 모형화할 때 많이 사용한다.

로그변환은 상관관계에 큰 영향을 미치는 않지만 연산의 의미가 크게 달라지므로 해석에 주의해야 한다.

### 탄력성, log-log 모형

log-log 모형에서 계수는 탄력성으로 해석할 수 있다.

$$\ln x = a_0 + a_1 \ln p_x + a_2 \ln p_y + a_3 \ln m + \varepsilon$$

$$\frac{d\ln x}{d\ln p} = \frac{dx/x}{dp/p} = \frac{dx}{dp} \frac{p}{x}$$

### 증가율, log-linear 모형

경제학에선 관심 변수의 변화를 증가율로 측정하는 경우가 많다. <br>
보통 가격이 얼마 올랐느냐가 아니라 몇 % 증가했는지 관심을 갖으며 <br>
총국내생산 등 경제 지표들의 변화를 이야기할 때도 증가율을 자주 사용한다.

log-log 모형에선 계수가 로그의 밑 값에 영향을 받지 않지만, <br>
log-linear 모형에선 자연로그를 사용한다.

log-linear 모형에서 계수는 percentage 변화를 의미한다.

$$\frac{ \Delta Y }{Y} \approx \frac{d \ln Y}{d t} = \frac{d Y / Y}{dt} $$

경제학에서 이야기하는 정태상태에선 모든 변수들의 증가율이 일정하다. <br>
따라서 이런 경우에는 증가율을 사용하면 변수들의 관계를 선형관계로 분석할 수 있다.

### linear-log 모형

해당 변수가 $k$배 커지면 종속변수는 $\beta \log( k )$ 만큼 증가한다.

$$ \beta \left\{ \log\left( k \cdot x \right) - \log\left( x \right) \right\}
= \beta \log\left( \frac{k \cdot x}{x} \right) = \beta \log( k )$$

<span style="color:navy; font-size: 1.25em; font-weight: bolder;"> 음압의 측정 Decibel, dB </span>

소음의 척도로 사용하는 데시벨은 상용로그를 사용한 비율이다. 기본 음압을 인간 청각의 한계인 $p_{\scriptsize{\text{ref}}} =$ 20 micropascals 로 설정하고 (0데시벨) 기본 음압과의 비율에 상용로그를 취한 값에 20을 곱해 사용한다.

$$20 \log_{10} \frac{p}{p_{\scriptsize{\text{ref}}}} \text{dB}$$

6데시벨마다 음량은 2배가 된다.

<span style="color:navy; font-size: 1.25em; font-weight: bolder;"> 로그변환에 사용하는 로그의 밑 </span>

탄력성의 크기는 로그의 밑에 영향을 받지 않지만
log-linear모형이나 linear-log모형에선 추정계수를 해석할 때 밑의 크기를 고려해야 한다.

$$\frac{d \, \log y }{ d\, x} = \frac{dy/dx}{y \ln 10}$$


[](https://ahvalnews.com/sites/default/files/inline-images/covid-confirmed-cases-since-100th-case.png)

(10, 1,000)과 (20, 10,000)의 두번째 값에 상용로그를 취하면 (10, 3)과 (20, 4)가 된다.<br>
기울기는 1/10 = 0.1 이다.

$$\frac{d \, \log y }{ d\, x} = \frac{dy/dx}{y \ln 10} = \frac{dy/dx}{y \times 2.30} = 0.10$$

따라서 증가율은 $2.30 \times 0.10 = 0.23$이고 $72\div 23 = 3.04$이므로 약 3일마다 2배씩 증가한다.

````python
x = data.index[3:]
X = sm.add_constant(x)
y = data.ln_income[3:]

model = sm.OLS(y, X).fit()

fig, ax = plt.subplots(figsize=(10,6))

data.ln_income[3:].plot(ax=ax)
model.fittedvalues.plot(ax=ax)

ax.set_xticks(range(1933, 1952))
ax.set_xticklabels(range(1933, 1952), rotation="vertical", fontsize = 14)
ax.set_xlabel(None)
ax.set_ylabel("Log Disposable Income", fontsize = 16)
ax.set_ylim(5.6, 7.6)

string = f"average slope btw 1933-51: {model.params[1]:.2f}"
ax.text(1935, 7.1, string, fontsize=16)
ax.grid(axis="y")
plt.suptitle("Figure 2. ln (disposable Income)", fontsize = 18);
#plt.savefig("fig1_disposable_income.png", dpi=600);
````

## 선형회귀모형 linear regression

- 선형회귀모형은 종속변수와 독립변수 사이에 선형관계를 가정한다.
- 독립변수의 수에 따라 단순회귀모형과 다중회귀분석으로 구분한다.
- 선형회귀모형으로는 인과관계를 설명하지 못하므로 해당 분야의 이론모형이 필요하다.


- $\mathbf{x}=(\mathbf{x}_1,\cdots,\mathbf{x}_K)$: features, attributes, <br>독립변수, 설명변수, independent variable, explanatory variables, <br>covariates, regressors


- $\mathbf{y}$: target, 종속변수, dependent variable


- 두 변수 사이에 선형관계를 가정하고 둘 사이의 관계를 추정하는 모형

$$\mathbf{y} = \beta_0 + \beta_1 \mathbf{x}_1 + \beta_2 \mathbf{x}_2 + \cdots + \beta_K \mathbf{x}_K + \underbrace{\beta_{K+1} \mathbf{x}^*_{K+1} + \beta_{K+2} \mathbf{x}^*_{K+2} + \cdots + \color{gray}{\text{non-deterministic part}}}_{\text{not observable}}$$

$$\mathbf{y}_i = \beta_0 + \sum^{K}_{k=1} \beta_k \mathbf{x}_{ik} + \varepsilon_i, \quad y_i = X_i \beta + \varepsilon_i$$
<br><br>





### 선택의 기준

- Mean Squared Error (MSE, L2)
    - Linear least squares estimation
    - Ordinary Least Square, Weighted Least Square, Generalized Least Square
- Likelihood funciton
    - Maximum Likelihood estimation 

### 모수의 추정

- 실제값과 추정량과의 편차 제곱을 가장 작게하는 모수를 선택
- mean squared error (MSE) 혹은 residual sum of squares (RSS)를 극소화

| Old | Popular  | calculation |
| :-- | :-- | :-- |
| SSR, sum of squres due to regression | ESS, explained sum of squares | $$\sum^n_{i=1} \left( \hat{y}_i - \bar{y} \right)^2$$ |
| SSE, sum of suqres of errors/residuals | RSS, residual sum of squares | $$\sum_{i=1}^n   (y_i-\hat{y}_i)^2$$ |
| SST, total sum of squares | TSS, total sum of squares | $$\sum_{i=1}^n   (y_i-\bar{y})^2$$ |

<br>

````python
def point_to_fitted_line(x, y, degree=1, fitted_line=True, segment=True, ax=None, 
                         face_color="lightgray", edge_color="dimgray",
                         line_color="gray", index=None):
    '''
    show the distance between data point and fitted line
    can use polyfit
    '''
    
    if ax is None: ax = plt.gca()
        
    slope, intercept = np.polyfit(x, y, deg=degree)
    f = lambda x: intercept + slope*x
    
    x_unit = (np.max(x) - np.min(x))/50
    y_unit = (np.max(y) - np.min(y))/50
    
    xlim=(np.min(x) - 10*x_unit, np.max(x) + 10*x_unit)
    ylim=(np.min(y) - 10*y_unit, np.max(y) + 10*y_unit)
    

    if fitted_line:
        ax.plot(xlim, [f(x) for x in xlim], c=line_color, 
                lw=2, label="fitted line")
    ax.scatter(x,y, ec=edge_color, c=face_color, s=20, label="sample")            
    ax.set(xlim=xlim, ylim=ylim)
        
    for i in range(len(y)):
        
        # data point to the fitted line
        label="residual/error" if i == 1 else ""
        if segment: ax.plot([x[i],x[i]], [y[i], f(x[i])], "r-", lw = 1.2, label=label)
        
        if not isinstance(index, type(None)):
            if y[i] >= f(x[i]):
                ax.annotate(index[i], xy=(x[i] - x_unit, y[i] + y_unit), fontsize=12, clip_on=True)
            else: ax.annotate(index[i], xy=(x[i] - x_unit,y[i] - 3*y_unit), fontsize=12, clip_on=True)
````

````python
teen_birth = pd.read_csv("https://github.com/k5yi/econ2005/blob/master/datasets/teen_birth.txt?raw=True", sep="\t")

random_sampling = np.random.RandomState(1).randint(1, len(data), size = 8)
teen_birth = teen_birth.loc[random_sampling]

x = teen_birth.PovPct.to_numpy()
y = teen_birth.Brth15to17.to_numpy()
state = teen_birth.Location.to_numpy()

fig, ax = plt.subplots(figsize=(8,6))
point_to_fitted_line(x, y, index=state, ax=ax)

ax.axhline(y = np.mean(y), c="lightgray", lw=1, label=r"mean of $y$")
ax.set_xlabel(r"$x$", fontsize=14)
ax.set_ylabel(r"$y$", fontsize=14, rotation=0)
ax.legend(fontsize=14, scatterpoints=3)
plt.suptitle("Figure 3. Minimize the sum of squared deviations", fontsize=18)
plt.tight_layout()
````

## Gauss-Markov Theorem

$$ y_i = \beta_0 + \beta_1 x_i + \varepsilon_i$$

$$\begin{eqnarray}
\hat{\beta}_0 &=& \bar{y} - \hat{\beta}_1 \bar{x}= \bar{y} - \frac{\sigma_{xy}}{\sigma^2_x} \bar{x} \\
\hat{\beta}_1 &=& \frac{\sigma_{xy}}{\sigma^2_x}
\end{eqnarray}$$

모수의 추정값을 계산하고 통계적인 해석하기 위해선 모집단과 표본에 대한 가정이 필요하다.

- $K\ge 2$일 경우

$$\begin{eqnarray}
\hat{\beta}_{\scriptsize{\text{OLS}}} &=& \arg \min_\beta \sum^n_{i=1} 
\left(y_i- X_i \beta\right)^2 \\
&=& \arg \min_\beta \left( \mathbf{y}- X\beta \right)^\intercal \left( \mathbf{y}- X\beta \right)
\end{eqnarray}$$
<br>
- OLS estimator
<br><br>
$$ \hat{\beta}_{\scriptsize{\text{OLS}}} = (X^\intercal X)^{-1} X^\intercal y, \quad
\text{Var}(\hat{\beta}_{\scriptsize{\text{OLS}}} | X) = \hat{\sigma}^2(X^\intercal X)^{-1} $$

### condition for existence

1\. no perfect collinearity, variations in $x$
$$ \text{Var}(x) \ne 0$$
<br><br>
$$ \hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}, \quad \hat{\beta}_1 = \frac{\sigma_{xy}}{\sigma^2_x} $$

###  condition for **unbiasedness**

$$E(\hat{\beta}_0) =\beta_0, \quad E(\hat{\beta}_1) =\beta_1$$

2\. linear relationship<br>
$$y_i = \beta_0 + \beta_1 x_i + \varepsilon_i$$
<br>
3\. random sampling<br>
$$(y_i, x_i) \sim \text{ IID }$$
<br>
4\. zero conditional mean (strict exogeneity, conditional independence)<br>
$$E(\varepsilon_i|x_i) = 0$$<br>

````python
def draw_gaussian_at(support, sd=1.0, height=1.0, 
        xpos=0.0, ypos=0.0, ax=None, **kwargs):
    if ax is None:
        ax = plt.gca()
    gaussian = np.exp((-support ** 2.0) / (2 * sd ** 2.0))
    gaussian /= gaussian.max()
    gaussian *= height
    return ax.plot(gaussian + xpos, support + ypos, **kwargs)
````

````python
np.random.seed(42)
sample_size = 100

# data generating process
f = lambda x: 1+1.2*x
x = np.random.randint(1, 9, sample_size)
disturbances = np.random.randn(sample_size)*1.5
y = f(x) + disturbances

fig, ax = plt.subplots(2, 1, figsize=(8,8), sharex=True)
ax=ax.flatten()

# plot data points
for i in range(len(x)):
    draw_gaussian_at(np.linspace(-6, 6), sd=1.5, height = 0.6,
                     xpos=x[i], ypos=f(x[i]), ax=ax[0], lw=.2, color="lightgray")
    draw_gaussian_at(np.linspace(-6, 6), sd=1.5, height = 0.6,
                     xpos=x[i], ypos=0, ax=ax[1], lw=.2, c="lightgray")

# plot fitted lines 
point_to_fitted_line(x, y, index=None, degree=1, segment=False, ax=ax[0],
                     face_color="skyblue", edge_color="blue", line_color="lightgray")

ax[1].axhline(y=0, c="lightgray")
point_to_fitted_line(x, disturbances, index=None, degree=1, segment=False, ax=ax[1], fitted_line=False,
                     face_color="pink", edge_color="crimson", line_color="lightgray")

# decors of the axes
y_ticks_ = np.unique(x)[::2]
[ax[0].plot([0,x], [f(x), f(x)], c="dimgrey", lw=.5) for x in y_ticks_]
y_label = [f"$E(y|x=${i}$)$" for i in y_ticks_]
ax[0].set(xticks=x, xlim=(0, 9), yticks=f(y_ticks_))
ax[0].set_yticklabels(y_label, fontsize=14)    

ax[0].legend(loc="lower right", fontsize=14, scatterpoints=3, ncol=2)
ax[0].grid(axis="x")

ax[1].legend(loc="upper right", fontsize=14, scatterpoints=3, ncol=2)
ax[1].grid(axis="x")

ax[0].set_title(r"$E(y_i |x_i) = \beta_0 + \beta_1 x_i$", fontsize=16)
ax[1].set_title(r"$E(\varepsilon_i|x_i) = E(y_i |x_i) - \left( \beta_0 + \beta_1 x_i \right)$", fontsize=16)

plt.suptitle("Figure 5. Conditional Distribution of Targets", fontsize=18)
plt.tight_layout()
````

## 추정 결과의 해석


- 귀무가설을 기각하지 못하는 것은 기각하는데 필요한 충분한 표본이 없기 때문이다.
- 귀무가설을 기각하는 이유는 채택하는데 필요한 충분한 표본이 없기 때문이다.



### 추정계수

- 추정량은 종속변수와 독립변수 사이의 공분산 혹은 상관계수, Pearson's correlation coefficient
<br><br>
$$ y_i = \beta_0 + \beta_1 x_i + \varepsilon_i $$
<br><br>
$$ \hat{\beta}_1 = \frac{\text{Cov}(x,y)}{\text{Var}(x)} $$
<br>
만일 $\sigma^2_x = \sigma^2_y=1$라면,
<br>
$$ \hat{\beta}_1 = \rho_{x,y} $$
<br>
- 상관계수는 선형정도를 나타내지만 상관계수로 그 정도를 짐작하기는 불가능하다.
- 시각화는 발표에서뿐 아니라 자료분석과정에서 중요한 역할을 한다.
<br><br>

### 가설검증 - Null Hypothesis Significance Testing

- Significance test
$$\text{H}_0: \beta_k = 0$$

1. 유의수준 결정
2. 귀무가설이 참일 경우 추정량 $\hat{\beta}_k$의 분포
3. 표본통계량보다 극단적인 값이 발생할 확률과 유의수준을 비교하여 기각여부를 판단
  - 단측검증: $\Pr\left(E(\hat{\beta}_k) < 0 \right)$ 혹은 $\Pr\left(E(\hat{\beta}_k) > 0 \right)$
  - 양측검증: $\Pr\left(E(\hat{\beta}_k) < 0  \text{ or } E(\hat{\beta}_k) > 0 \right)$

- Constraint test
$$\text{H}_0: f(\beta) = 0$$
Wald test, Log-likelihood ratio test
<br><br>

### 결정계수, R-squared, coefficient of determination

- 귀무가설 하의 분포 (추정값)과 실제 분포 사이의 관계
- 추정값과 실제값 사이의 상관계수의 제곱
- 계수값의 유의성이 주 관심이라면 결정계수의 크기는 크게 문제가 되지 않는다.
- 전체 변동성중 설명이 가능한 변동성의 비율
<br><br>

<span style="color:navy; font-size:1.5em; font-weight:bolder">결정계수와 상관계수</span>

- 두 개념 모두 인과관계와 아무 관계가 없다.

$$\text{corr}^2(y, \hat{y}) = \text{R}^2$$
<br>
$$\begin{eqnarray}
\rho^2 (y, \hat{y}) &=& \frac{\text{Cov}^2(y, \hat{y})}{ \text{Var}(y) \text{Var}(\hat{y}) } 
= \frac{\text{Cov}^2(\hat{y}+e, \hat{y})}{ \text{Var}(y) \text{Var}(\hat{y}) } \\
&=& \frac{ \left( \text{Var}(\hat{y}) + \text{Cov}(\hat{y},e) \right)^2 }{ \text{Var}(y) \text{Var}(\hat{y}) } 
= \frac{\text{Var}^2(\hat{y})}{ \text{Var}(y) \text{Var}(\hat{y}) } \\
&=& \frac{\text{Var}(\hat{y})}{ \text{Var}(y) } = \frac{\text{RSS}}{\text{TSS}} = 1 - \frac{\text{ESS}}{\text{TSS}} = \text{R}^2 \\
\end{eqnarray}$$

<br>

<span style="color:navy; font-size:1.5em; font-weight:bolder">결정계수와 multicollinearity</span>

- pairwise coefficient of determination
  - 두 변수사이의 결정계수가 0.6이라 하면 관련 다중공선성 문제는 거의 무시
  - 0.6에서 0.8 사이라면 모형을 손보거나 조심해서 해석
  - 0.8 이상이라면 제외시키는 것을 고려
  - 0.9 이상이라면 특별한 이유가 없는 한 제외
<br>
<br>
- pairwise correlation coefficient
  - $\text{corr}^2(x_1, x_2) = \text{R}^2_{x_1,x_2}$
<br>
<br>
- 다중회귀분석에선 variance inflation factor를 참고
<br>
$$\text{VIF}_k = \frac{1}{1-\text{R}^2_{x_k, x_{-k}}}$$
<br>
- 2.5, 5, 10이 상관계수 0.6, 0.8, 0.9에 해당

##  non-zero conditional mean 과 bias

- 수요와 공급 모형에서 배웠듯이 공급곡선에 대한 정보를 얻기 위해서는 다음 조건을 만족해야 한다.
  1. 표본기간 중 탄력성은 동일하게 유지한 채 수요곡선이 충분히 변해야 한다.
  2. 가격과 공급량 사이의 관계를 추정하기 위해서는 두 변수 이외의 다른 변수들은 '고정'되어 있어야 한다. 이 조건은 공급에 영향을 미치는 다른 변수들을 통제하여 만족시킬 수 있다.
  3. 통계적인 추론을 위해서는 OLS의 기본가정이 만족되어야 한다.
  

- 수요/공급 추정에서 가장 곤란한 문제는 균형가격과 균형량의 내생성이다.
- 내생성 문제는, $E(X_i \cdot \varepsilon_i) \ne 0$, 주로 측정오차나 변수누락, 그리고 연립방정식 모형에서 발생한다.
<br><br>
$$\begin{eqnarray}
E(\beta_{\scriptsize{\text{OLS}}}) = \beta_1 + \frac{\text{Cov}(x,\varepsilon)}{\text{Var}(x)}
\end{eqnarray}$$
<br>
- (원하는 효과) 공급곡선을 추정할 때는 (공급계획에 영향을 미치지 않는) 수요의 변화로 인한 가격변화 <br>
- (불필요한 효과) 공급곡선을 추정할 때는 공급계획의 변화로 인한 가격변화

$$\begin{eqnarray}
E(\hat{\beta}_1) &=& \frac{\text{Cov}(x,y)}{\text{Var}(x)} \\
&=& \frac{\text{Cov}(x, \beta_0 + \beta_1 x + \varepsilon)}{\text{Var}(x)} \\
&=& \frac{\beta_1 \text{Var}(x) + \text{Cov}(x, \varepsilon)}{\text{Var}(x)} \\
&=& \beta_1 + \frac{\text{Cov}(x, \varepsilon)}{\text{Var}(x)}
\end{eqnarray}$$

- **Omitted variable**
  - when a regression equation leaves out one or more relevant variables which are **correated** with covariates
- Measurement error

- **Endogeneity, Simultaneity**

### Structural Equation

- Covariate이 내생변수

$$\begin{eqnarray}
q^s &=& \beta p + u \\ 
q^d &=& \gamma p + v \\ 
q^s &=& q^d
\end{eqnarray}$$
- 시장균형 및 표본
<br>
$$ p^* = \frac{v - u}{\beta-\gamma}, \quad q^* = \frac{\gamma v - \beta u}{\beta-\gamma}$$
<br>
- $u \perp v$ 이라면
<br>
$$E(p^* \cdot u) = E\left(\frac{v - u}{\beta-\gamma} \cdot u\right) 
= \frac{E(v \cdot u)}{\beta-\gamma} - \frac{E(u^2)}{\beta-\gamma} 
= - \frac{\sigma^2_u}{\beta-\gamma} \ne 0$$
<br><br>

### 시장자료와 다중회귀분석

- 공급곡선을 식별하기 위해서는 해당 곡선 상의 변화만 고려해야 한다.
- 가격 이외의 변화로 인한 영향을 제거하기 위해 통제변수 사용
<br><br>
$$\ln q = a_0 + a_1 \ln p + a_2 \ln r + a_3 \ln w $$
<br>
가격 이외에 생산량에 영향을 미치는 요인들을 통제하면 (효과를 제거하면) 공급곡선의 이동으로 발생하는 효과를 제거할 수 있다. <br>
따라서 수요곡선의 이동에 의해 발생한 가격변화만을 고려할 수 있다.
<br><br>

````python
f = lambda a,b,c: a+b*c

def eq(supply_a, supply_b, demand_a, demand_b):
    q = (demand_a - supply_a)/(supply_b-demand_b)
    return q, f(supply_a, supply_b, q)
````

### 이윤극대화

- 하나의 작물을 얼마나 재배할지 결정하는 문제
<br><br>
$$ \max_q  p\cdot q - c(q, z)$$
<br>
F.O.C.
$$\begin{eqnarray}
&p &- \,\text{MC}(q^*, z) = 0 \\
&p & = \text{MC}(\bar{q}, z) + \text{MC}'(\bar{q}, z) \cdot (q^*-\bar{q}), \quad \bar{q}\in [0, q^*] \\
&q^* & = \frac{p - \text{MC}(\bar{q}, z)}{\text{MC}'(\bar{q}, z)} + \bar{q} = \alpha_0 + \alpha_1 p
\end{eqnarray}$$
<br>
S.O.C.
<br>
$$ - \text{MC}'(q^*, z) < 0 $$

### 비용함수와 재배면적 제약

- 여러 개의 작물대안 중 하나를 선택는 문제
<br><br>
어떤 작물을 심던 심어야 한다면 각 작물의 판매가격이 가장 중요한 변수가 되므로 <br>
비용함수의 세부적인 형태는 별다른 문제가 되지 않을 수 있다.

$$\begin{array}{rl}
\max_{x,y} & p_x x + p_y y - c(x, y, z) \\
\text{subject to }& x + y = 1
\end{array}$$

1계조건을 선형근사하면,
<br>
$$\begin{eqnarray}
p_x - \,\text{C}_x (x^*, y^*, z) &=& p_x - \left\{ \text{C}_x (\bar{x}, y^*, z) + \text{C}_{xx} (\bar{x}, y^*, z) \cdot (x^*-\bar{x}) \right\} = 0 \\
p_y - \,\text{C}_y (x^*, y^*, z) &=& p_y - \left\{ \text{C}_y (x^*, \bar{y}, z) + \text{C}_{yy} (x^*, \bar{y}, z) \cdot (y^*-\bar{y}) \right\}= 0 
\end{eqnarray}$$
<br>
여기서 $\text{C}^\prime, \text{C}^{\prime\prime} > 0$로 가정하고 (정확한 2계필요조건은 좀 더 복잡하다) $x^*$에 대해 정리하면,
<br><br>
$$\begin{array}{ll}
p_x - p_y  -  a_{11} - a_{12} (x^*-\bar{x}) + a_{21} + a_{22} (y^*-\bar{y}) = 0, \;\;& a_{ij} >0 \\
p_x - p_y  -  a_{11} - (a_{12} + a_{22}) x^* + a_{12}\bar{x} + a_{21} + a_{22} (1- \bar{y}) = 0&
\end{array}$$
<br>
따라서 회귀식은 다음과 같이 정리할 수 있다.
$$\begin{eqnarray}
x^* &=& \frac{p_x - p_y }{ a_{22} + a_{12} } + a_{12}\bar{x} - a_{22} \bar{y} + A \\
x^* (p_x, p_y, z) &=& \alpha_0 + \alpha_1 p_x + \alpha_2 p_y
\end{eqnarray}$$

### 공급(계획)

- 교재에서 사용한 Suits의 모형과 변수는 다음과 같다.

$$ \ln q_t = \alpha_0 + \alpha_1 \ln p_{t-1} + \alpha_2 \ln p^c_{t-1} 
+ \alpha_3 \ln p^v_{t-1} + \alpha_4 d^c + \alpha_5 d^w + \varepsilon_t$$

| 변수   |   설명
|:--    | :--
| $q$   | 수확이 가능한 수박 재배량 (백만)
| $p$   | 수박의 산지 평균판매가격 (천개)
| $p^c$ | 면화의 파운드 당 연평균 수입
| $p^v$ | 일반적인 야채의 산지평균판매가격 (index)
| $d^c$ | ln_cotton_price acreage allotment program dummy, 1934-51
| $d^w$ | 이차세계대전 dummy, 1943-46

````python
data.loc[:,"ln_price_lagged"] = data.loc[:, "ln_price"].shift()
data.loc[:,"ln_cotton_lagged"] = data.loc[:, "ln_cotton"].shift()
data.loc[:,"ln_vegi_lagged"] = data.loc[:, "ln_vegi"].shift()
data.loc[:,"ln_wage_lagged"] = data.loc[:, "ln_wage"].shift()

data.loc[:, "cotton_acreage_program"] = ((data.index >= 1934) & (data.index <= 1951)).astype(int)
data.loc[:, "world_war_2"] = ((data.index >= 1943) & (data.index <= 1946)).astype(int)

data.columns
````

````python
x_supply = data.loc[:, ['ln_price_lagged', 'ln_cotton_lagged', 'ln_vegi_lagged',
                        "cotton_acreage_program", "world_war_2"]]
X_supply = sm.add_constant(x_supply)
y = data.loc[:, "ln_crop"]

supply = sm.OLS(y, X_supply, missing="drop")
supply_fit = supply.fit()
print(supply_fit.summary())
````

### 가격과 생산량 사이의 관계

- 추정한 수요/공급곡선과 실제 자료를 비교하기 위해 통제 변수들의 효과를 제거한 관찰값을 계산하여 도표에 그려보자.

- 즉, $\ln p_{t-1}$ 과 다음 값의 관계는 가격과 공급량 평면에 그린 공급곡선이다.

$$ \ln q_t -( \hat{\alpha}_1 + \hat{\alpha}_3 p^c_{t-1} 
+ \hat{\alpha}_4 p^v_{t-1} + \hat{\alpha}_5 d^c_t + \hat{\alpha}_6 d^w_t)$$

````python
b_hat = supply_fit.params.copy()
b_hat["ln_price_lagged"] = 0
b_hat["const"] = 0

q_observed = data.loc[:,['ln_crop']].to_numpy().ravel()
q_supplied_controlled = q_observed - (b_hat*x_supply).sum(axis=1)

p = np.linspace(2, 7, 100)
q_hat = supply_fit.params["const"] + supply_fit.params["ln_price_lagged"] * p

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(q_observed, data.loc[:,['ln_price']], "ob", 
        alpha = .4, markeredgecolor = "navy",
        markersize = 12, label = "Data")
ax.plot(q_supplied_controlled, data.loc[:,['ln_price']], "or",
        alpha = .4, markeredgecolor = "crimson",
        markersize = 12, label = "Controlled Quatity Supplied")
ax.plot(q_hat, p, "grey", label = "Estimate of supply")
ax.legend(loc = "lower right", fontsize = 16)

ax.set_xlabel("ln(Quantity) (millions)", fontsize = 14)
ax.set_ylabel("ln(Price) (US $ per 1,000 melons)", fontsize = 14)

fig.suptitle('Figure 13. Esitmation of Supply Curve')
fig.tight_layout()
````

### 이윤극대화 문제

$$ L(h,L; \, \lambda, \eta) = ph-wL - \lambda(h-AL^a) -\eta(h-q)$$
First order condition
$$\begin{array}{llcc}
h : & \;\;\; p - \lambda^*  -\eta^* &=& 0 \\
L : & - w + \lambda^* a A (L^*)^{a-1} &=& 0 \\
\lambda : & - \left\{ h^*-A(L^*)^a \right\} &=& 0\\
\eta : & \;\;\;\eta^*(h^*-q^*) &=& 0
\end{array}$$
<br>
$0 <a <1$이라면,
<br>
$$ \ln h^* = \ln A + \frac{a}{1-a} \left\{ \ln A + \ln a + \ln \left(\frac{p}{w} + \eta^* \right) \right\} \le \ln q^*$$

$$ \ln h^* = \ln A + \frac{a}{1-a} \left\{ \ln A + \ln a + \ln \left(\frac{p}{w} + \eta^* \right) \right\} \le \ln q^*$$
<br>
따라서, $h^* < q$ 일 경우 $\eta^* =0$이므로
$$\ln h^* = \beta_0 + \beta_1 \ln \left(\frac{p}{w} \right)$$
<br>
만일 $h^* = q$라면  $\eta^* > 0$는 단위당 이윤으로 해석할 수 있다. 선형근사 값을 구해보면
<br><br>
$$ \ln \left( \frac{p}{w} + \eta^* \right) \approx \ln \left( \frac{p}{w} \right) + \left( \frac{w}{p} \right) \eta^*$$

### 실질가격의 의미

$$
\begin{array}{ll}
\max_{h, L}& p h - w L \\
\text{subject to }& h = A L^a\\
&h \le q
\end{array}
\quad\quad = \quad\quad 
\begin{array}{ll}
\max_{h, L}& \frac{p}{w} h - L \\
\text{subject to }& h = A L^a\\
&h \le q
\end{array}
$$
<br><br>
일반적으로
$$
\begin{eqnarray}
\max_{h, L, K} &\;\;& ph - wL - rK\\
\text{subject to }&& h = A L^a K^b\\
&&h \le q
\end{eqnarray}
$$
<br>
일반적으로 이윤은 $p/w$에 대해 1차동차가 아니며, 따라서 요소수요 역시 $p/w$에 대해 0차동차가 아니다.

### Suits' model

$$\ln h_t = \beta_0 + \beta_1 \ln \left(\frac{p_t}{w_t} \right) + \beta_2 \ln q_t + \varepsilon_{2t}$$

- spurious relationship의 가능성이 높다.
<br>
- Suits의 분석에서 '실질가격'을 사용한 이유는 자유도 1개를 살리려는 목적이었다.

### $h_t < q_t$ 자료를 이용한 Suits 모형 추정

- $t \in \left\{ t | h_t < q_t \right\}$에 대해 다음 회귀식을 추정한다.
<br><br>
$$\begin{eqnarray}
\ln h_t &=& \beta_1 + \beta_2 \ln \frac{p_t}{w_t} + \beta_3 \ln q_t + \varepsilon_{2t}\\
\ln h_t &=& \beta_1 + \beta_2 \ln \frac{p_t}{w_t} + \varepsilon_{2t}
\end{eqnarray}$$
<br>
여기서 $h_t$는 실제 수확량, $w_t$는 South Atlantic States의 임금지수이다.<br>
<br>
- Stewart의 사례분석에서는 전체 표본을 사용한 결과로 보이며<br>
Suits는 수확량과 재배량 제약이 binding하지 않은 표본만 사용하였다.

````python
data.loc[:, "ln_real_price"] = data.loc[:, "ln_price"]-data.loc[:,"ln_wage"]
data.loc[:, "ln_real_price_lagged"] = data.loc[:, "ln_real_price"].shift(1)
````

````python
x_harvest = data.loc[:, ["ln_real_price", "ln_crop"]]

X_harvest = sm.add_constant(x_harvest)
y = data.ln_harvest

index = data.ln_harvest < data.ln_crop

harvest = sm.OLS(y[index], X_harvest[index], missing="drop")
harvest_fit = harvest.fit()
print("Suits Model: h < q")
print(harvest_fit.summary())
````

## 수확량의 추정과 문제점

- 위 추정 결과에 따르면 가격은 수확량에 통계적으로 의미있는 영향을 미친다고 보기 어렵다.
- $h_t$가 종속변수일 때 $p_t$를 설명변수로 사용할 경우 문제점
<br><br>
$$ \color{red}{\ln h_t} = \beta_1 + \beta_2 \ln \frac{\color{red}{p_t}}{\color{green}{w_t}} + \beta_3 \ln q_t + \varepsilon_{2t} $$
<br>
1. Identification
 - 공급곡선을 식별하는 것이므로 $q_t$는 공급곡선을 이동시키는 원인이어야 한다.<br>
 - 추정식이 공급곡선인지 수요곡선인지 이론적인 설명이 필요하다.


2. endogeneity
 - $h_t$와 $p_t$는 균형에서 결정되는 내생변수들이다.

### 2 Stage Least Squares Regression Analysis

- $X_i = Z_i \gamma + \eta$에서 문제를 일으키는 $\eta$ 부분을 제거
<br><br>
- First stage estimation
<br>
$$ X_i = \gamma_0 + \gamma_1 Z_i + \gamma_2 W_i + \nu$$
<br>

 - 여러 개의 도구변수를 사용하거나 상관 관계를 크게하기 위해 사용
 - $W_i$는 원래 모형의 회귀방정식에서 사용한 독립변수들을 포함할 수 있다.


- Second stage estimation
<br>
$$ y_i = \beta_0 + \beta_1 \hat{X}_i + \beta_2 X_{-i} + e = \beta \hat{X} + e$$
<br>
$\hat{X} = Z(Z^\intercal Z)^{-1} Z^\intercal X$이므로 $\hat{X} = P X$라 하자.<br>
$P$는 symmetric한 idempotent 행렬이므로,
<br>
$$\begin{eqnarray}
\hat{\beta}_{\scriptsize{\text{2SLS}}} &=& (\hat{X}^\intercal \hat{X})^{-1} \hat{X}^\intercal y \\
&=& \left[ (P X)^\intercal P X \right] ^{-1} (P X)^\intercal y  \\
&=& \left[ X^\intercal P X \right] ^{-1} X^\intercal P y  
\end{eqnarray}$$

- 도구변수의 조건
  - 필요한 것은 가격에 직접적인 영향을 미치면서 공급량에는 영향을 미치는 않는 변수이다. <br>
  - 즉 공급곡선을 이동시키는 요인을 배재해야 한다. <br>
  - 수확철에 폭풍이 발생한다면 공급곡선 이동을 통해 가격과 생산산량에 모두 영향을 미친다.
<br><br>
$$\begin{array}{ll}
\text{condition 1 : } & \text{NONE} \Rightarrow (\text{instrument}, h_t) \\
\text{condition 2 : } & \text{instrument} \Rightarrow p_t \Rightarrow h_t \\
\text{condition 3 : } & \text{instrument} {\not}{\Longrightarrow} h_t \\
\text{condition 4 : } & \text{corr}(\text{instrument}, h_t) \text{ is significantly different from zero.} 
\end{array}$$

$$ \color{red}{\ln h_t} = \beta_1 + \beta_2 \ln \frac{\color{red}{p_t}}{w_t} + \beta_3 \ln q_t + \varepsilon_{2t} $$

- $p_{t-1}$은 수확량 결정 이전에 결정되는 변수로 수확량에 직접적인 영향을 미친다고 보기 어렵다.
<br><br>
- $p_{t-1}$가 $q_t$에 미치는 영향이 절대적이므로 $q_t$를 통한 영향이 있을 수 있지만 
  - $h_t$에 직접적인 영향을 미치지 않고,
  - $q_t$를 포함시킬 경우 다중공선성 이외에는 별 문제가 없다.
<br><br>
- $p_t$와 $p_{t-1}$의 상관계수는 0.926으로 아주 크다.

[Testing for weak instruments](https://en.wikipedia.org/wiki/Instrumental_variables_estimation): The strength of the instruments can be directly assessed because both the endogenous covariates and the instruments are observable. A common rule of thumb for models with one endogenous regressor is: the F-statistic against the null that the excluded instruments are irrelevant in the first-stage regression should be larger than 10.

### Hausman test

$$\text{H}_0 : \text{The 'endogenous' variable has no endogenity.}$$

- Hausman test<br>
Hausman test는 2SLS의 first stage 추정결과를 이용하여<br>
다음 회귀식에서 $\gamma$의 추정값으로 $\ln \frac{p_t}{w_t}$의 내생성 여부를 확인한다.
$$ \ln h_t = \beta_1 + \beta_2 \ln \frac{p_t}{w_t} + \gamma \widehat{ \ln \frac{p_t}{w_t} } + \beta_3 \ln q_t + \varepsilon_{2t} $$
여기서 $\widehat{ \ln \frac{p_t}{w_t} }$은 다음 회귀식의 fittedvalue이다.
$$\ln \frac{p_t}{w_t} = \delta_0 + \delta_1 \ln \frac{p_{t-1}}{w_{t-1}} + \nu_t$$
<br>
- 검증결과의 해석
  - 만일 도구변수 고유의 정보가 종속변수와 유의미한 상관관계가 없다면 내생성이 없다고 판단한다.<br>
  - 검증결과는 도구변수의 선택에도 영향을 받으므로 해석에 주의한다.<br>
  - 따라서 Hausman test의 귀무가설은 다음과 동일하다.
$$H_0: \gamma = 0$$

### 내생성검증결과

$$\begin{array}{rl}
\text{H}_0: &\text{ 추정에 내생성 문제가 없다. }\\
\text{H}_0: &\text{ 선택한 도구변수로는 내생성이 있다고 할 수 없다. }
\end{array}$$

### Wald test

LR test statistic의 근사값으로 LR 통계량에 비해 계산이 복잡하지만 <br>
회귀방정식을 하나만 추정하면 검증을 할 수 있다는 장점이 있다. <br>
하나의 모수에 대한 통계량은 $t$-값의 제곱이고,  <br>
모수의 선형, 비선형 조합(1차근사)에 대한 검증이 가능하다.

$$ H_0 : R \theta = r $$

$$ \left( R\hat{\theta} - r \right)^\intercal \left[ 
R \hat{\Sigma}_\theta R^\intercal
\right]^{-1}
\left( R\hat{\theta} - r \right) \sim \chi^2_Q$$


예를 들어 두 개의 계수의 합이 0인지 검증할 때 $R = [1 1]$ 을 사용한다. 이떄 통계량은

$$ 
\begin{eqnarray}
&\left( \begin{bmatrix} 1 & 1 \end{bmatrix}
\begin{bmatrix} \hat{\theta}_1 \\ \hat{\theta}_2 \end{bmatrix} \right)^\intercal
\left(
\begin{bmatrix} 1 & 1 \end{bmatrix}
\begin{bmatrix}
\hat{\sigma}^2_1 & \hat{\sigma}_{12} \\
\hat{\sigma}_{21}  & \hat{\sigma}^2_2 
\end{bmatrix}
\begin{bmatrix} 1 \\ 1 \end{bmatrix} 
\right)^{-1}
\left( \begin{bmatrix} 1 & 1 \end{bmatrix}
\begin{bmatrix} \hat{\theta}_1 \\ \hat{\theta}_2 \end{bmatrix} \right) \\
&= 
\frac{\left( \hat{\theta}_1 + \hat{\theta}_2 \right)^2
}{ \left(\hat{\sigma}^2_1+\hat{\sigma}^2_2+2\hat{\sigma}_{12} \right)
}
\sim \chi^2_{(2)}
\end{eqnarray}$$

### 선택 모형

- OLS 추정값을 사용하여 수요함수를 계산한다.

<br>
$$ \ln p_t = \gamma_0 + \gamma_1 \ln \frac{h_t}{n_t} + \gamma_2 \ln \frac{y_t}{n_t} 
+ \gamma_4 p^f_t + \varepsilon_{3t} $$

$$ \ln p_t = -3.4878 + -0.8441 \ln \frac{h_t}{n_t} 1.5618 \ln \frac{y_t}{n_t} 
-0.7846 p^f_t + \varepsilon_{3t} $$
<br>

- 추정 후 경제적으로 의미있는 해석을 위해 위식을 수요함수로 정리한다.

$$ \ln \frac{h_t}{n_t} = - \frac{\hat{\gamma}_0}{\hat{\gamma}_1}
+ \frac{1}{\hat{\gamma}_1} \ln p_t - \frac{\hat{\gamma}_2}{\hat{\gamma}_1} \ln \frac{y_t}{n_t}
 - \frac{\hat{\gamma}_3}{\hat{\gamma}_1} \ln p^f_t$$

- 추정계수값을 이용하여 정리하면 다음의 추정식을 얻을 수 있다.

$$\ln \frac{h_t}{n_t} = -4.13  -1.18 \ln p_t +1.85 \ln \frac{y_t}{n_t}  -0.93 \ln p^f_t$$


- Stewart는 3SLS 모형을 선택하였다.

$$\ln \frac{h_t}{n_t} = \hat{\beta}_0  -1.125 \ln p_t + 1.75 \ln \frac{y_t}{n_t}  - 0.968 \ln p^f_t$$

$$ \ln p_t = \hat{\gamma}_0  - 0.8891 \ln \frac{h_t}{n_t} + 1.5563 \ln \frac{y_t}{n_t}  - 0.8604 \ln p^f_t$$

````python
def variance(a, b, est, cov):
    # a is the index of the variable in numerator, b the denominator
    
    if a == None:
        var = cov.loc[b,b]/est[b]**4
    
    else:
        m1, m2 = est.loc[[a,b]]
        v1, cov_, _, v2 = cov.loc[[a,b], [a,b]].to_numpy().flatten()
        var = ( 1/m2**4 ) * ( m2**2 * v1 - 2*m1*m2*cov_  + m1**2 * v2 )
    
    return var

est = inv_demand.params
cov = inv_demand.cov

s_constant = np.sqrt(variance('const', 'consumption', est, cov))
s_price_elasticity = np.sqrt(variance(None, 'consumption', est, cov))
s_income_elasticity = np.sqrt(variance('ln_income', 'consumption', est, cov))
s_cross_elasticity = np.sqrt(variance('ln_freight', 'consumption', est, cov))

std = f"({s_constant:.2f}) & ({s_price_elasticity:.2f}) & ({s_income_elasticity:.2f}) & ({s_cross_elasticity:.2f})"
print(std)
````

역수요함수를 추정하면 수요함수 계수의 정확한 분포를 알기 어렵고, 따라서 관련한 추론이 쉽지 않다. <br>
보통 1차 혹은 2차 Taylor series apporximation 을 사용한다.  <br>
$\mu = (\mu_X, \mu_Y)$라 하자. 불편추정량이라면 1차 근사값은 다음과 같다.

$$\begin{eqnarray}
E(f(X,Y)) &\approx& E \left\{ f(\mu) + f_1(\mu) (X - \mu_X) + f_2(\mu) (Y - \mu_Y) \right\} \\ 
&=& E \left\{ f(\mu) \right\} 
+ E \left\{ f_1(\mu) (X - \mu_X) \right\} 
+ E \left\{ f_2(\mu) (Y - \mu_Y) \right\} \\
&=& E \left\{ f(\mu) \right\} 
+ f_1(\mu) E \left\{ (X - \mu_X) \right\} 
+ f_2(\mu) E \left\{ (Y - \mu_Y) \right\} \\
&=& f(\mu) = f(\mu_X, \mu_Y)
\end{eqnarray}$$

$$\begin{eqnarray}
\text{Var}(f(X,Y)) &\approx& 
E \left[ \left\{ f(\mu) + f_1(\mu) (X - \mu_X) + f_2(\mu) (Y - \mu_Y) \color{red}{- f(\mu)} \right\}^2 \right] \\ 
&=& 
E \left[ \left\{ f_1(\mu) (X - \mu_X) + f_2(\mu) (Y - \mu_Y) \right\}^2 \right] \\
&=& 
E \left[
f^2_1(\mu) (X - \mu_X)^2 + 2 f_1(\mu)(X - \mu_X) f_2(\mu)(Y - \mu_Y)
+ f^2_2(\mu) (Y - \mu_Y)^2 
\right] \\
&=& f^2_1(\mu)\text{Var}(X) + 2 f_1(\mu)f_2(\mu)\text{Cov}(X,Y) + f^2_2(\mu)\text{Var}(Y)
\end{eqnarray}$$

$$\begin{eqnarray}
\text{Var}\left(\frac{a}{b}\right) \approx 
\frac{\mu^2_a}{\mu^2_b} 
\left\{
\left( \frac{\sigma_a}{\mu_a} \right)^2  + 2 \frac{ \sigma_{ab} }{\mu_a\mu_b} + \left( \frac{\sigma_b}{\mu_b} \right)^2
\right\}
\end{eqnarray}$$

$$\begin{eqnarray}
\text{Var}\left(\frac{a}{b}\right) \approx 
\frac{\mu^2_b \sigma^2_a - 2 \mu_a\mu_b \sigma_{ab}  + \mu^2_a \sigma^2_b}{\mu^4_b}
\end{eqnarray}$$

- 역수요함수의 추정결과를 수요함수로 변환

````python
est = inv_demand.params
cov = inv_demand.cov

constant = -est[0]*price_elasticity
price_elasticity = 1/est[3]
income_elasticity = -est[1]*price_elasticity
cross_elasticity = -est[2]*price_elasticity

string = r"\ln \frac{h_t}{n_t} ="
string += f" {constant:+.2f}  {price_elasticity:+.2f} "
string += r"\ln p_t "
string += f"{income_elasticity:+.2f} "
string += r"\ln \frac{y_t}{n_t} "
string += f" {cross_elasticity:+.2f} "
string += r"\ln p^f_t"

s_constant = np.sqrt(variance('const', 'consumption', est, cov))
s_price_elasticity = np.sqrt(variance(None, 'consumption', est, cov))
s_income_elasticity = np.sqrt(variance('ln_income', 'consumption', est, cov))
s_cross_elasticity = np.sqrt(variance('ln_freight', 'consumption', est, cov))

std = f"&({s_constant:.2f}) && ({s_price_elasticity:.2f}) && ({s_income_elasticity:.2f}) && ({s_cross_elasticity:.2f})&"
print(string)
print(std)
````
