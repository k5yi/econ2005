adstrd_code가 동일하지 않아 100배해주는 함수

```python
def times100(data, col):
    for i in data.index:
        val=data.at[i, col]
        val=val*100;
        data.loc[i,col]=val
```

```python
fashion_pre_regulation = fashion_need[(fashion_need.date >= 20200216) & (fashion_need.date <= 20200229)]
fashion_aft_regulation = fashion_need[(fashion_need.date >= 20200302) & (fashion_need.date<= 20200315)]

print(fashion_pre_regulation)
print()
print(fashion_aft_regulation)
```

```python
fashion_pre_regulation_sales = fashion_pre_regulation.groupby(['adstrd_nm','signgu_nm']).sum()['sale'].reset_index()
fashion_pre_regulation_sales.rename(columns = {"sale": "sum of sale"}, inplace = True)
fashion_pre_regulation_sales
```

```python
fashion_aft_regulation_sales = fashion_aft_regulation.groupby(['adstrd_nm','signgu_nm']).sum()['sale'].reset_index()
fashion_aft_regulation_sales.rename(columns = {"sale": "sum of sale"}, inplace = True)
fashion_aft_regulation_sales
```

```python
differ_regulation_sales=fashion_pre_regulation_sales.merge(fashion_aft_regulation_sales, how='inner', on='adstrd_nm')
differ_regulation_sales
```

```python
differ_regulation_sales=differ_regulation_sales.drop([differ_regulation_sales.index[242], differ_regulation_sales.index[243]])
differ_regulation_sales
```

```python
sinsa = differ_regulation_sales[(differ_regulation_sales.adstrd_nm == '신사동') ]
sinsa
```

```python
differ_regulation_sales.rename(columns = {"signgu_nm_x": "signgu_nm"}, inplace = True)
differ_regulation_sales.rename(columns = {"sum of sale_x": "sum_before"}, inplace = True)
differ_regulation_sales.rename(columns = {"sum of sale_y": "sum_after"}, inplace = True)
differ_regulation_sales=differ_regulation_sales.drop('signgu_nm_y',axis=1)

differ_regulation_sales
```

```python
differ_regulation_sales['sale_change'] = differ_regulation_sales.sum_after - differ_regulation_sales.sum_before
differ_regulation_sales['changed%'] = differ_regulation_sales.sale_change / differ_regulation_sales.sum_before
differ_regulation_sales
```
