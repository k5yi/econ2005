```python
#4.19일인 이유: 4.19부터 사회적 거리두기 완화, 생활속거리두기로 전환 
#6월 중순부터 다시 거리두기가 강회되기떄문에 우리데이터와 부합하는면이 많음
#질병관리본부홈페이지 에서 대규모 집단감염발생을 2.18~5.5로 명시한 자료가있음
#사회적 거리두기라는 말이 브리핑에 사용되는건 3월부터지만, 3.1 이후부터 집단감염, 지역사회 감염에 우려가 보임



fpopl_before2['adstrd_code']=fpopl_before2['adstrd_code'].astype(str)
fpopl_before2['adstrd_code'] = fpopl_before2['adstrd_code'].apply(lambda x:x[:5])
fpopl_before2=fpopl_before2.groupby(by=['adstrd_code'],as_index=False).popltn_cascnt.mean()


fpopl_after2['adstrd_code']=fpopl_after2['adstrd_code'].astype(str)
fpopl_after2['adstrd_code']=fpopl_after2['adstrd_code'].apply(lambda x:x[:5])
fpopl_after2=fpopl_after2.groupby(by=['adstrd_code'],as_index=False).popltn_cascnt.mean()
```

```python
add_col=[]
for i in range(len(fpopl_after2)):
    add_col.append(fpopl_after2['popltn_cascnt'][i])

fpopl_diff2=fpopl_before2
fpopl_diff2['covid_after']=add_col
fpopl_diff2.columns=['adstrd_code','covid_before','covid_after']

#유동인구 차이값 계산
fpopl_diff2['diff']=fpopl_diff2['covid_before']-fpopl_diff2['covid_after']
fpopl_diff2=fpopl_diff2.sort_values(by=['diff'],ascending=False)
fpopl_diff2=fpopl_diff2.reset_index()
fpopl_diff2=fpopl_diff2.drop(['index'],axis=1)
```

```python
#지역 위도 경도 표시
region=pd.read_csv('C:\\Users\\tagom\\Region.csv',usecols=['province','city','latitude','longitude'])
region_seoul=region[region['province']=='Seoul']
region_seoul['province']='Seoul'
region_seoul['total_ads']=region_seoul['province']+' '+region_seoul['city']
region_s=region_seoul[['total_ads','latitude','longitude']]

region_s=region_s[1:]
region_s=region_s.reset_index()
region_s=region_s.drop(['index'],axis=1)
region_s=region_s.sort_values(by=['total_ads'],ascending=False)
region_s=region_s.reset_index()
region_s=region_s.drop(['index'],axis=1)

ad_lat=[]
ad_lon=[]
for i in range(len(region_s)):
    ad_lat.append(region_s['latitude'][i])
for i in range(len(region_s)):
    ad_lon.append(region_s['longitude'][i])

fpopl_diff['latitude']=ad_lat
fpopl_diff['longitude']=ad_lon

fpopl_diff=fpopl_diff.sort_values(by=['diff'],ascending=False)
```
